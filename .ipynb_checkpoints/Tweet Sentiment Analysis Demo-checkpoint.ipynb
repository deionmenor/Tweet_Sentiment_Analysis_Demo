# Tweet Sentiment Analysis Demonstration

The following demonstration will show you a step-by-step process on how you can perform sentimental analysis on tweet data

The following libraries are needed for this demo:

````python
import tweepy          
import pandas as pd     
import numpy as np

from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

## Setting up the Twitter Application

Before anything else, we need to make our own Twitter application so that our Tweepy library will be able to interact with the website's API. A quick tutorial can be found [here](http://docs.inboundnow.com/guide/create-twitter-application/)

In our main python file, we will make a section for the required keys/tokens for using the API. Substitute the empty strings below with the appropriate keys. Remember to never upload these keys in a public repo, they are secret for a reason. Anyone with knowledge of these keys can possibly manipulate your account!

````python
######################################################################
# Consume:
CONSUMER_KEY    = ''
CONSUMER_SECRET = ''

# Access:
ACCESS_TOKEN  = ''
ACCESS_SECRET = ''

########################################################################
````

````python
# API's setup:
def twitter_setup():
    """
    Utility function to setup the Twitter's API
    with our access keys provided.
    """
    # Authentication and access using keys:
    auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
    auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)

    # Return API with authentication:
    api = tweepy.API(auth)
    return api
````

## Getting the Tweets

We now call the ```twitter_setup``` function to make our api object. The Twitter API limits the number of tweets you can receive per request, so we request for the account's tweets multiple times until we get at least 1000 tweets. For this demonstration, we will be looking at ABS-CBN News Channel's Twitter [@ANCALERT](https://twitter.com/ANCALERTS)

The snippet below will also iteratively print out how many tweets it received so far.

````python

api = twitter_setup()
target_user = "ANCALERTS"
tweets = []

    #make initial request for most recent tweets (200 is the maximum allowed count)
new_tweets = api.user_timeline(screen_name = target_user,count=200)

    #save most recent tweets
tweets.extend(new_tweets)

    #save the id of the oldest tweet less one
oldest = tweets[-1].id - 1

    #keep grabbing tweets until there are no tweets left to grab
while len(tweets) < 1000:

        #all subsiquent requests use the max_id param to prevent duplicates
    new_tweets = api.user_timeline(screen_name = target_user,count=200,max_id=oldest)

        #save most recent tweets
    tweets.extend(new_tweets)

        #update the id of the oldest tweet less one
    oldest = tweets[-1].id - 1

    print("...{num} tweets downloaded so far".format(num=len(tweets)))

````

Printing out the length of ```tweets``` gives us exactly 1000 tweets to work with, as shown below

````python
print("{} Tweets extracted from {}.\n".format(len(tweets),target_user))
````

## Using a DataFrame

Next, we will be using a pandas ```DataFrame```, which will allow us to more easily edit and transform the table that holds our tweet data. The code below creates a DataFrame ```data``` that first only has a column holding the text attribute of our tweets. 
We then add more columns to our DataFrame to hold other possibly relevant information, such as the date the tweet was posted and the number of favorites and retweets

````python
data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['tweets'])

## Make new columns in our DataFrame
data['len']  = np.array([len(tweet.text) for tweet in tweets])
data['ID']   = np.array([tweet.id for tweet in tweets])
data['Date'] = np.array([tweet.created_at for tweet in tweets])
data['Likes']  = np.array([tweet.favorite_count for tweet in tweets])
data['RTs']    = np.array([tweet.retweet_count for tweet in tweets])
````

We can use the ```head``` method to peek at the first five entries of our DataFrame. The table shows the most recent tweets from the account. 

````python
display(data.head())
````

We can use ```numpy``` to calculate the average tweet length of the account

````python
mean = np.mean(data['len'])
m = "{0:.4g}".format(mean)
print("{} has an average tweet length of {} characters\n".format(target_user,m))
````

We can use ```numpy```'s max function to get the the tweets with the maximum number of likes and retweets. 

````python
fav_max = np.max(data['Likes'])
rt_max = np.max(data['RTs'])

## finds the tweet with the maximum number of tweets and saves the row number
fav = data[data.Likes == fav_max].index[0]
rt  = data[data.RTs == rt_max].index[0]
````

In this case, it turned out that the same tweet had the most number of likes and the most number of retweets.

````python
print("{}'s most favorited tweet is: \n {}".format(target_user,data['tweets'][fav]))
print("Number of favorites: {}\n\n".format(fav_max))

print("{}'s most retweeted tweet is: \n {}".format(target_user, data['tweets'][rt]))
print("Number of retweets: {}".format(rt_max))
````

We can also try plotting how the tweets are received through time. The code below converts our tweet data into a  ```pandas``` time series object, which we can then plot with ```matplotlib```


````python
tfav = pd.Series(data=data['Likes'].values, index=data['Date'])
tret = pd.Series(data=data['RTs'].values, index=data['Date'])
````

````python
tfav.plot(figsize=(16,10), label="Likes", legend=True)
tret.plot(figsize=(16,10), label="Retweets", legend=True)
````

## Sentiment Analysis

Let's do some sentiment analysis! We will use ```textblob```, a easy-to-use Python library that serves as an API for various common natural language processing tasks such as noun phrase extraction, sentiment analysis, classification, translation, and more. We will also be importing the ```re``` library ( which allows you to use regular expressions) in order to clean our tweet text data.

````python
from textblob import TextBlob
import re

def clean_tweet(tweet):
    '''
    Utility function to clean the text in a tweet by removing 
    links and special characters using regex.
    '''
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())

def analyze_sentiment(tweet):
    '''
    Utility function to classify the polarity of a tweet
    using textblob.
    '''
    analysis = TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity > 0:
        return analysis.sentiment.polarity
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return analysis.sentiment.polarity
````

We use the ```analyze_sentiment``` function above to create another column for our DataFrame called SA. This column indicates whether or not a certain text is classified by its "polarity" as positive, negative, or neutral.

````python
# We create a column with the result of the analysis:
data['SA'] = np.array([ analyze_sentiment(tweet) for tweet in data['tweets'] ])

# We display the updated dataframe with the new column:
display(data.head(5))
````

We will now look at the percentage of positive, neutral, and negative tweets from the user

````python
pos_tweets = [ tweet for index, tweet in enumerate(data['tweets']) if data['SA'][index] > 0]
neu_tweets = [ tweet for index, tweet in enumerate(data['tweets']) if data['SA'][index] == 0]
neg_tweets = [ tweet for index, tweet in enumerate(data['tweets']) if data['SA'][index] < 0]
````

````python
print("Percentage of positive tweets: {}%".format(len(pos_tweets)*100/len(data['tweets'])))
print("Percentage of neutral tweets: {}%".format(len(neu_tweets)*100/len(data['tweets'])))
print("Percentage de negative tweets: {}%".format(len(neg_tweets)*100/len(data['tweets'])))
````

Graphing the results using ```matplotlib```, we can see that the number of neutral texts are very much ahead of the other two categories

````python
objects = ('Positive', 'Neutral', 'Negative')
y_pos = np.arange(len(objects))
performance = [len(pos_tweets),len(neu_tweets),len(neg_tweets)]

fig_size = [12,9]

plt.rcParams["figure.figsize"] = fig_size
 
plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.ylabel('Sentiment')
plt.title('Sentimental Analysis of @{} Tweets'.format(target_user))
plt.figure()
````

This scatterplot seems to suggest that all of @ANCALERTS's tweets with over 300 likes have a neutral sentiment result. However, this may be a biased result considering over 58% of the tweets are considered neutral
